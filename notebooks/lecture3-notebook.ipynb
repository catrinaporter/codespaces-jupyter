```python
evens = 0
odds = 0
```

```python
value = random.randint(1000,9999)
print(value)

if value % 2 == 1:
    print("value is odd")
    odds = odds + 1
else:
    evens += 1
print("value is", value)
print("odds", odds)
print("evens", evens)
```

```python
count = 0

while count < 50:
    value = random.randint(1000,9999)
    print(value)

    if value % 2 == 1:
        print("value is odd")
        odds = odds + 1
    else:
        evens += 1
    print("value is", value)
    print("odds", odds)
    print("evens", evens)
    count += 1
```
